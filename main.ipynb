```python
from chains.extraction_chain import extraction_chain
from chains.matching_chain import matching_chain
from chains.scoring_chain import scoring_chain
from chains.explanation_chain import explanation_chain

print("Chains Loaded")
```

```python
with open("data/job_description.txt") as f:
    job_desc = f.read()

print("Job Description Loaded")
```

```python
def run_pipeline(resume_text):

    # Step 1: Extraction
    extracted = extraction_chain.invoke({
        "resume": resume_text
    })

    # Step 2: Matching
    matched = matching_chain.invoke({
        "resume_data": extracted,
        "job_desc": job_desc
    })

    # Step 3: Scoring
    score = scoring_chain.invoke({
        "match_data": matched
    })

    # Step 4: Explanation
    explanation = explanation_chain.invoke({
        "resume_data": extracted,
        "match_data": matched,
        "score": score
    })

    return extracted, matched, score, explanation
```

```python
files = ["strong_resume.txt", "average_resume.txt", "weak_resume.txt"]

for file in files:
    print("\n" + "="*50)
    print(f"Processing: {file}")

    with open(f"data/{file}") as f:
        resume = f.read()

    extracted, matched, score, explanation = run_pipeline(resume)

    print("\n--- Extracted ---")
    print(extracted)

    print("\n--- Matched ---")
    print(matched)

    print("\n--- Score ---")
    print(score)

    print("\n--- Explanation ---")
    print(explanation)
```
